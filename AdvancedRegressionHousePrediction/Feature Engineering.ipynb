**For numeric datasets if they have large number of outliers Nan values should be replaced with median or mode for better accuracy rather than with mean**

```python
def fill_nan_num(df,num_feat):
    for feature in num_feat:
        median=df[feature].median()
        df[feature+'nan']=np.where(df[feature].isnull(),1,0)
        df[feature].fillna(median,inplace=True)
fill_nan_num(df,numeric_feature_with_nan)
df[numeric_feature_with_nan].isnull().sum()
```

### Handling temporal features

**In EDA we saw that the numeric data was skewed i.e. it was not normally distributed so in order to make it normally distributed we will perform log transformation but only on those that don't have any 0 value in their column**

```python
num_features=['LotFrontage','LotArea','1stFlrSF','GrLivArea']
for feature in num_features:
    df[feature]=np.log(df[feature])

df.head()
```

```python
test_num_features=['LotFrontage','LotArea','1stFlrSF','GrLivArea']
for feature in test_num_features:
    df_test[feature]=np.log(df_test[feature])

df_test.head()
```

## Handling rare categorical feature

**These are records that represent less than 1% of the observations. For example if a feature has 3 categories A, B and C and if category B has less than 1% of observation we change that record's category to some other category as it wont have a major effect on our target feature.**

```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
categorical_features
```

```python
test_categorical_features=[feature for feature in df_test.columns if df_test[feature].dtype=='O']
test_categorical_features
```

```python
for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'RareVar')
```

```python
for feature in test_categorical_features:
    temp=df_test.groupby(feature)['LotArea'].count()/len(df_test)
    temp_df=temp[temp>0.01].index
    df_test[feature]=np.where(df_test[feature].isin(temp_df),df_test[feature],'RareVar')
```

## Feature Scaling

**We have to do feature scaling as there are numeric features which have different units so scaling is needed for better accuracy.**

```python
features_scale=[feature for feature in df.columns if df[feature].dtypes!='O' and feature not in ['Id','SalePrice']]
```

```python
test_features_scale=[feature for feature in df_test.columns if df_test[feature].dtypes!='O' and feature not in ['Id','SalePrice']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[features_scale])
```

```python
scaler1=MinMaxScaler()
scaler1.fit(df_test[test_features_scale])
```

```python
X_train=pd.concat([df[['Id','SalePrice']+categorical_features].reset_index(drop=True),pd.DataFrame(scaler.transform(df[features_scale]),columns=features_scale)],axis=1)
X_train.set_index('Id')
```

```python
X_train.to_csv('X_train.csv',index=False)
```

```python
X_test=pd.concat([df_test[['Id']+test_categorical_features].reset_index(drop=True),pd.DataFrame(scaler1.transform(df_test[test_features_scale]),columns=test_features_scale)],axis=1)
X_test.set_index('Id')
```
