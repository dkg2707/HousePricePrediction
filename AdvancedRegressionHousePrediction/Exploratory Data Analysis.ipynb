```python
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt
%matplotlib inline

#To display all columns
pd.pandas.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('train.csv')

print(df.shape)
```

```python
df.head()
```

# Exploratory data analysis is gather information from data so comparing all features with the dependent features i.e. target feature SalePrice

**Get percentage of missing values for columns that have missing values**

```python
features_with_na=[feature for feature in df.columns if df[feature].isnull().sum()>1]
feature_na_dict={}
for feature in features_with_na:
    feature_na_dict[feature]=np.round(df[feature].isnull().mean(),4)
print("% of missing values")
print(feature_na_dict)
```

```python
for feature in features_with_na:
    data=df.copy()
    #If feature had missing value its Nan is converted to one else 0 this is done to see if missing values have any impact on our target variable i.e. salesprice
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

**We can see from graphs that the feature that have more Nan values have higher Saleprice median so there is clearly a relation between them.**

# Numeric variables

```python
features_with_num=[feature for feature in df.columns if df[feature].dtypes!='O']


print('Number of Numerical features '+str(len(features_with_num)))

df[features_with_num].head()
```

**This numeric variables also contains temporal variables(i.e. date time variables) like YearBuilt, YearRemodAdd, GarageYrBlt, YrSold (4 variables). These are called temporal variables as with time new data will generate. We will have to take difference between yearblt,yrsold and more operations on this variables as columns and add them to our dataframe to get meaningful information.**

```python
year_feature=[feature for feature in features_with_num if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
for feature in year_feature:
    print(feature, df[feature].unique())
```

```python
df.groupby('YrSold')['SalePrice'].median().plot(kind='line',x='Year Sold',y='Median House Price',title='House Price vs Year Sold')
```

**As the year is increasing the prices are decreasing which logically cant be True. So it is safe to assume that there is no direct relationship between SalePrice and yearsold. Hence we have to perform some operations on temporal variables**

## Working with Temporal variables
**Here we are comparing difference between yearsold and other features w.r.t Saleprice**

```python
for feature in year_feature:
    if feature!='YrSold':
        data=df.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel('YrSold - '+feature)
        plt.ylabel('SalePrice')
        plt.show()
```

**We can clearly see that as the age of house increases the price decreases.**

```python
#Discrete and continous numeric features

discrete_feature=[feature for feature in features_with_num if len(df[feature].unique())<25 and feature not in year_feature+['Id']]
print('No. of discrete features: '+str(len(discrete_feature)))
print(discrete_feature)
```

```python
# Finding Relationship between discrete features and Target Variable: SalePrice

for feature in discrete_feature:
    data=df.copy()
    data.groupby(feature)['SalePrice'].median().plot(kind='bar',x=feature,y='SalePrice',title=feature)
    plt.show()
```

**For some features like OverallQual there is exponential rise in median of saleprice. So we can safely say that there is a realtionship between discrete variable and Target Variable**

```python
#Continous Variables
continous_features=[feature for feature in features_with_num if feature not in discrete_feature + year_feature+['Id']]
print('No. of continous features: '+str(len(continous_features)))
print(continous_features)
```

```python
for feature in continous_features:
    data=df.copy()
    #for continous variables we use histograms to check if they are normally distributed or not
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

**Each distribution has different distribution so no data is normally distributed. Hence to make them normally distributed we will perform log transformations.**

```python
for feature in continous_features:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

### Outliers

```python
for feature in continous_features:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

**There are many outliers we will later remove them**
